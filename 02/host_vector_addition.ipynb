## The `vecadd` Function

This function does the actual work. A few C++ concepts to note:

- **`float*`** — a *pointer* to a float. Arrays in C++ are passed as pointers to their first element, so `A_h` here represents the whole array.
- **`int n`** — the number of elements; we pass this explicitly because the function can't know the array size on its own.
- The loop runs once per element, adding the corresponding entries of A and B and storing the result in C.

```c++
// Compute vector sum C_h = A_h + B_h
void vecadd(float* A_h, float* B_h, float* C_h, int n) {
    for (int i = 0; i < n; ++i) {
        C_h[i] = A_h[i] + B_h[i];
    }
}
```

## Memory Allocation

In C++, `new float[N]` allocates a block of memory on the **heap** large enough to hold `N` floats and returns a pointer to it.

> Unlike a regular variable (which lives on the *stack* and is automatically cleaned up), heap memory stays alive until you explicitly free it with `delete[]`. We do that at the end.

```c++
int N = 1000; // number of elements — change as needed

// Allocate three arrays of N floats on the heap
float* A = new float[N];
float* B = new float[N];
float* C = new float[N];
```

## Filling Arrays with Random Values

- **`srand(42)`** seeds the random number generator with `42`. Using a fixed seed means you get the same "random" numbers every run — useful for reproducibility.
- **`rand()`** returns a random integer between `0` and `RAND_MAX` (a large constant, typically 32767 or 2147483647).
- **`static_cast<float>(...) / RAND_MAX`** converts it to a float in the range `[0.0, 1.0]`.

```c++
srand(42); // seed for reproducibility
for (int i = 0; i < N; ++i) A[i] = static_cast<float>(rand()) / RAND_MAX;
for (int i = 0; i < N; ++i) B[i] = static_cast<float>(rand()) / RAND_MAX;
```

## Running the Addition and Printing Results

We call `vecadd`, then print the first 10 results to verify it worked.

`std::cout` is C++'s standard way to print to the console. The `<<` operator chains values together — think of it as "send this to the output stream".

```c++
vecadd(A, B, C, N);

std::cout << "Result C = A + B (first 10 elements):\n";
for (int i = 0; i < 10; ++i)
    std::cout << "C[" << i << "] = " << C[i] << "\n";
```

## Freeing Memory

Every `new` must be paired with a `delete[]`. Forgetting this causes a **memory leak** — the program holds onto RAM it no longer needs until it exits.

```c++
delete[] A;
delete[] B;
delete[] C;
```
