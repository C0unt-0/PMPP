```c++
#include <iostream>
#include <cstdlib>

// Compute vector sum C_h = A_h + B_h
void vecadd(float* A_h, float* B_h, float* C_h, int n) {
    for (int i = 0; i < n; ++i) {
        C_h[i] = A_h[i] + B_h[i];
    }
}


int N = 1000; // number of elements — change as needed

// Memory allocation for arrays A, B, and C
float* A = new float[N];
float* B = new float[N];
float* C = new float[N];

// I/O to read A and B, N elements each
std::cout << "Enter " << N << " values for A:\n";
for (int i = 0; i < N; ++i) std::cin >> A[i];

std::cout << "Enter " << N << " values for B:\n";
for (int i = 0; i < N; ++i) std::cin >> B[i];

vecadd(A, B, C, N);

std::cout << "Result C = A + B:\n";
for (int i = 0; i < N; ++i)
    std::cout << "C[" << i << "] = " << C[i] << "\n";

delete[] A;
delete[] B;
delete[] C;

```
